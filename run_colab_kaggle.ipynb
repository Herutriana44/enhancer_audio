**Catatan:** Ganti `YOUR_REPO_URL` dengan URL repository Git Anda (contoh: `https://github.com/username/enhancer_audio.git`)

```python
# ========== KONFIGURASI ==========
# Ganti dengan URL repository Git Anda
REPO_URL = "https://github.com/Herutriana44/enhancer_audio.git"
PORT = 5000
```

```python
# ========== 1. GIT CLONE ==========
import subprocess
import os
subprocess.run(["git", "clone", REPO_URL, "enhancer_audio"], check=True)
os.chdir("enhancer_audio")
print("Working dir:", os.getcwd())
```

```python
# ========== 2. INSTALL DEPENDENCIES ==========
# Install FFmpeg (untuk Colab/Kaggle)
!apt-get update -qq && apt-get install -y -qq ffmpeg

# Install Python packages
!pip install -q flask flask-cors demucs audio-denoiser noisereduce torch torchaudio ffmpeg-python soundfile numpy

print("Dependencies installed!")
```

```python
# ========== 3. INSTALL NGROK ==========
!pip install -q pyngrok

from pyngrok import ngrok

# Set auth token jika punya (dapatkan gratis di https://ngrok.com)
# ngrok.set_auth_token("YOUR_NGROK_TOKEN")

print("ngrok installed!")
```

```python
# ========== 4. JALANKAN FLASK + NGROK ==========
import threading
import time

def run_flask():
    from app import app
    app.run(host="0.0.0.0", port=PORT, debug=False, use_reloader=False)

# Start Flask di background
flask_thread = threading.Thread(target=run_flask)
flask_thread.daemon = True
flask_thread.start()

# Tunggu Flask siap
time.sleep(3)

# Buka tunnel ngrok
tunnel = ngrok.connect(PORT)
public_url = tunnel.public_url if hasattr(tunnel, 'public_url') else str(tunnel)
print(f"\n" + "="*50)
print("AUDIO ENHANCER SIAP!")
print(f"Buka URL ini di browser: {public_url}")
print("="*50)
```
